```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/amazon_sales_data.csv")
```

```python
df['Segment'] = df['B2B'].apply(lambda x: 'B2B' if x else 'B2C')
df.groupby('Segment')['Amount'].sum().plot(kind='bar', title="B2B vs B2C Sales")
plt.ylabel("Amount")
plt.show()
```

```python
df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%', title="Order Distribution: B2B vs B2C")
plt.ylabel('')
plt.show()
```

```python
avg_order_value = df.groupby('Segment')['Amount'].mean()
avg_order_value.plot(kind='bar', title="Average Order Value per Segment")
plt.ylabel("Avg Order Amount")
plt.show()
```
