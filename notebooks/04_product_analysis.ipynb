```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/amazon_sales_data.csv")
```

```python
top_categories = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
top_categories.head(10).plot(kind='bar', title="Top Categories by Sales", ylabel="Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
qty_by_cat = df.groupby('Category')['Qty'].sum().sort_values(ascending=False)
qty_by_cat.head(10).plot(kind='bar', title="Top Categories by Quantity Sold", ylabel="Quantity")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Size', order=df['Size'].value_counts().index)
plt.title("Product Size Distribution")
plt.xticks(rotation=45)
plt.show()
```
