```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/amazon_sales_data.csv")
df['Date'] = df['Date'].astype(str).str.strip()
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')
```

```python
sales_over_time = df.groupby(df['Date'])['Amount'].sum()
sales_over_time.plot(kind='line', figsize=(12, 6), title="Total Sales Over Time", ylabel="Amount (INR)")
plt.grid(True)
plt.show()
```

```python
monthly_sales = df.resample('M', on='Date')['Amount'].sum()
monthly_sales.plot(kind='bar', figsize=(12, 6), title="Monthly Sales", ylabel="Amount (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Fulfilment')['Amount'].sum().plot(kind='bar', title='Sales by Fulfilment Type', ylabel='Total Amount')
plt.show()
```
